```python
from PyQt5 import QtCore, QtGui, QtWidgets
from Firstanalysis import Ui_First

class Ui_MainWindow(object):
    def OnClick(self):
        self.window = QtWidgets.QMainWindow()
        self.ui= Ui_First()
        self.ui.setupUi(self.window)
        MainWindow.hide()
        self.window.show()
        
    def OnClick2(self):
        self.window = QtWidgets.QMainWindow()
        self.ui= Ui_Secondone()
        self.ui.setupUi(self.window)
        MainWindow.hide()
        self.window.show()
        
    def OnClick4(self):
        self.window = QtWidgets.QMainWindow()
        self.ui= Ui_Third()
        self.ui.setupUi(self.window)
        MainWindow.hide()
        self.window.show()
        
    def OnClick6(self):
        self.window = QtWidgets.QMainWindow()
        self.ui= Ui_Lastpage()
        self.ui.setupUi(self.window)
        MainWindow.hide()
        self.window.show()
        
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(1470, 835)
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
        self.label = QtWidgets.QLabel(self.centralwidget)
        self.label.setGeometry(QtCore.QRect(0, -40, 1471, 911))
        self.label.setObjectName("label")
        self.label_2 = QtWidgets.QLabel(self.centralwidget)
        self.label_2.setGeometry(QtCore.QRect(900, 30, 471, 151))
        self.label_2.setObjectName("label_2")
        self.pushButton = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton.setGeometry(QtCore.QRect(1010, 380, 311, 81))
        self.pushButton.setObjectName("pushButton")
        self.pushButton.clicked.connect(self.OnClick2)
        self.pushButton_2 = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton_2.setGeometry(QtCore.QRect(1030, 270, 261, 81))
        self.pushButton_2.setObjectName("pushButton_2")
        self.pushButton_2.clicked.connect(self.OnClick)
        self.pushButton_3 = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton_3.setGeometry(QtCore.QRect(1020, 500, 301, 81))
        self.pushButton_3.setObjectName("pushButton_3")
        self.pushButton_3.clicked.connect(self.OnClick4)
        self.pushButton_4 = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton_4.setGeometry(QtCore.QRect(990, 610, 391, 81))
        self.pushButton_4.setObjectName("pushButton_4")
        self.pushButton_4.clicked.connect(self.OnClick6)
        MainWindow.setCentralWidget(self.centralwidget)
        self.menubar = QtWidgets.QMenuBar(MainWindow)
        self.menubar.setGeometry(QtCore.QRect(0, 0, 1470, 36))
        self.menubar.setObjectName("menubar")
        MainWindow.setMenuBar(self.menubar)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)

        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "MainWindow"))
        self.label.setText(_translate("MainWindow", "<html><head/><body><p><img src=\":/newPrefix/housing.webp\"/></p></body></html>"))
        self.label_2.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\"><span style=\" font-size:48pt; font-weight:600;\">HOUSING ANALYSIS</span></p></body></html>"))
        self.pushButton.setText(_translate("MainWindow", "Housing Median Age by Ocean Proximity"))
        self.pushButton_2.setText(_translate("MainWindow", "House Value Rankings"))
        self.pushButton_3.setText(_translate("MainWindow", "Average Loan Amount by Month and Year"))
        self.pushButton_4.setText(_translate("MainWindow", "Ocean Proximity Values for High-Value Homes"))
import housing

class Ui_First(object):
    def OnClick1(self):
        self.window = QtWidgets.QMainWindow()
        self.ui= Ui_MainWindow()
        self.ui.setupUi(self.window)
        MainWindow.hide()
        self.window.show()
        
    def draw_bar_graph(self):
        # Create a QGraphicsScene to hold the bar graph
        scene = QtWidgets.QGraphicsScene()
        self.graphicsView.setScene(scene)

        # Execute the SQL query to retrieve the data for the bar graph
        query = """
            SELECT a.city, AVG(h.median_house_value) as median_house_value
            FROM HouseOwnership ho
            JOIN DimAddress a ON ho.add_id = a.add_id
            JOIN DimHousing h ON ho.housing_id = h.housing_id
            GROUP BY a.city
            ORDER BY median_house_value ASC
            LIMIT 10;
        """
        # Replace the connection details with your own
        from DATA225utils import make_connection
        conn = make_connection('centaurus_wh.ini')
        cursor = conn.cursor()

        cursor.execute(query)
        data = cursor.fetchall()

        max_value = max(row[1] for row in data)

        # Create a QBrush to fill the bars in the graph
        brush = QtGui.QBrush(QtGui.QColor(149, 175, 192))

        # Calculate the width of each bar
        bar_width = self.graphicsView.width() / len(data)

        # Loop through the data and add a bar for each value
        for i, row in enumerate(data):
            # Calculate the height of the bar
            bar_height = (row[1] / max_value) * self.graphicsView.height()

            # Create a QGraphicsRectItem to represent the bar
            rect = QtWidgets.QGraphicsRectItem(
                i * bar_width, self.graphicsView.height() - bar_height,
                bar_width, bar_height
            )

            # Set the brush for the bar
            rect.setBrush(brush)

            # Add the bar to the scene
            scene.addItem(rect)

            # Add the value label above the bar
            value_label = QtWidgets.QGraphicsTextItem(str(round(row[1], 2)))
            value_label.setPos(i * bar_width + bar_width / 2 - value_label.boundingRect().width() / 2, self.graphicsView.height() - bar_height - 20)
            scene.addItem(value_label)

            # Add x-axis label
            x_label = QtWidgets.QGraphicsTextItem(row[0])
            x_label.setPos(i * bar_width + bar_width / 2 - x_label.boundingRect().width() / 2, self.graphicsView.height() + 10)
            scene.addItem(x_label)

        # Add y-axis label
        y_label = QtWidgets.QGraphicsTextItem("Median House Value")
        y_label.setPos(-50, self.graphicsView.height() / 2 - y_label.boundingRect().height() / 2)
        y_label.setRotation(-90)
        scene.addItem(y_label)
        
        self.graphicsView.scale(0.8, 0.8)

        # Close the database connection
        conn.close()

    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(1470, 836)
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
        self.label = QtWidgets.QLabel(self.centralwidget)
        self.label.setGeometry(QtCore.QRect(460, 50, 571, 81))
        self.label.setObjectName("label")
        self.graphicsView = QtWidgets.QGraphicsView(self.centralwidget)
        self.graphicsView.setGeometry(QtCore.QRect(130, 170, 821, 641))
        self.graphicsView.setObjectName("graphicsView")
        self.pushButton = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton.setGeometry(QtCore.QRect(1090, 400, 113, 32))
        self.pushButton.setObjectName("pushButton")
        self.pushButton.clicked.connect(self.draw_bar_graph)
        self.pushButton_2 = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton_2.setGeometry(QtCore.QRect(1090, 470, 113, 32))
        self.pushButton_2.setObjectName("pushButton_2")
        self.pushButton_2.clicked.connect(self.OnClick1)
        MainWindow.setCentralWidget(self.centralwidget)
        self.menubar = QtWidgets.QMenuBar(MainWindow)
        self.menubar.setGeometry(QtCore.QRect(0, 0, 1470, 36))
        self.menubar.setObjectName("menubar")
        MainWindow.setMenuBar(self.menubar)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)

        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "MainWindow"))
        self.label.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\"><span style=\" font-size:36pt;\">House Value Rankings</span></p></body></html>"))
        self.pushButton.setText(_translate("MainWindow", "Load"))
        self.pushButton_2.setText(_translate("MainWindow", "Back"))
        
class Ui_Secondone(object):
    def OnClick3(self):
        self.window = QtWidgets.QMainWindow()
        self.ui= Ui_MainWindow()
        self.ui.setupUi(self.window)
        MainWindow.hide()
        self.window.show()
    
    def loaddata(self):
        selected_index = self.comboBox.currentIndex()
        selected_item = self.comboBox.itemText(selected_index)
        sql = """
                SELECT o.ocean_proximity_value,
                       CASE 
                         WHEN h.housing_median_age BETWEEN 1 AND 20 THEN '1-20' 
                         WHEN h.housing_median_age BETWEEN 21 AND 40 THEN '20-40' 
                         WHEN h.housing_median_age BETWEEN 41 AND 62 THEN '40-62' 
                       END AS age_range, 
                       COUNT(*) as count
                FROM HouseOwnership ho
                JOIN DimHousing h ON ho.housing_id = h.housing_id
                JOIN DimOceanProximity o ON ho.proximity_id = o.proximity_id
                WHERE o.ocean_proximity_value = '{}'
                  AND h.housing_median_age IS NOT NULL
                GROUP BY 
                       CASE 
                         WHEN h.housing_median_age BETWEEN 1 AND 20 THEN '1-20' 
                         WHEN h.housing_median_age BETWEEN 21 AND 40 THEN '20-40' 
                         WHEN h.housing_median_age BETWEEN 41 AND 62 THEN '40-62' 
                       END
                ORDER BY 
                       CASE 
                         WHEN h.housing_median_age BETWEEN 1 AND 20 THEN '1-20' 
                         WHEN h.housing_median_age BETWEEN 21 AND 40 THEN '20-40' 
                         WHEN h.housing_median_age BETWEEN 41 AND 62 THEN '40-62' 
                       END;
            """.format(selected_item)
        from DATA225utils import make_connection
        conn = make_connection('centaurus_wh.ini')
        cursor = conn.cursor()
        cursor.execute(sql)
        result = cursor.fetchall()
        self.tableWidget.setRowCount(0)
        self.tableWidget.setColumnCount(len(cursor.description))
        headers = [field[0] for field in cursor.description]
        self.tableWidget.setHorizontalHeaderLabels(headers)
        for row_number, row_data in enumerate(result):
            self.tableWidget.insertRow(row_number)
            for column_number, data in enumerate(row_data):
                self.tableWidget.setItem(row_number, column_number, QtWidgets.QTableWidgetItem(str(data)))
        conn.close()
        
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(800, 600)
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
        self.tableWidget = QtWidgets.QTableWidget(self.centralwidget)
        self.tableWidget.setGeometry(QtCore.QRect(190, 220, 371, 121))
        self.tableWidget.setObjectName("tableWidget")
        self.tableWidget.setColumnCount(3)
        self.tableWidget.setRowCount(3)
        item = QtWidgets.QTableWidgetItem()
        self.tableWidget.setVerticalHeaderItem(0, item)
        item = QtWidgets.QTableWidgetItem()
        self.tableWidget.setVerticalHeaderItem(1, item)
        item = QtWidgets.QTableWidgetItem()
        self.tableWidget.setVerticalHeaderItem(2, item)
        item = QtWidgets.QTableWidgetItem()
        self.tableWidget.setHorizontalHeaderItem(0, item)
        item = QtWidgets.QTableWidgetItem()
        self.tableWidget.setHorizontalHeaderItem(1, item)
        item = QtWidgets.QTableWidgetItem()
        self.tableWidget.setHorizontalHeaderItem(2, item)
        self.label = QtWidgets.QLabel(self.centralwidget)
        self.label.setGeometry(QtCore.QRect(70, 20, 721, 61))
        self.label.setObjectName("label")
        self.pushButton = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton.setGeometry(QtCore.QRect(200, 410, 113, 32))
        self.pushButton.setObjectName("pushButton")
        self.pushButton.clicked.connect(self.loaddata)
        self.pushButton_2 = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton_2.setGeometry(QtCore.QRect(470, 410, 113, 32))
        self.pushButton_2.setObjectName("pushButton_2")
        self.pushButton_2.clicked.connect(self.OnClick3)
        self.comboBox = QtWidgets.QComboBox(self.centralwidget)
        self.comboBox.setGeometry(QtCore.QRect(320, 140, 141, 26))
        self.comboBox.setObjectName("comboBox")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.label_2 = QtWidgets.QLabel(self.centralwidget)
        self.label_2.setGeometry(QtCore.QRect(140, 140, 141, 16))
        self.label_2.setObjectName("label_2")
        MainWindow.setCentralWidget(self.centralwidget)
        self.menubar = QtWidgets.QMenuBar(MainWindow)
        self.menubar.setGeometry(QtCore.QRect(0, 0, 800, 36))
        self.menubar.setObjectName("menubar")
        MainWindow.setMenuBar(self.menubar)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)

        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "MainWindow"))
        item = self.tableWidget.verticalHeaderItem(0)
        item.setText(_translate("MainWindow", "New Row"))
        item = self.tableWidget.verticalHeaderItem(1)
        item.setText(_translate("MainWindow", "New Row"))
        item = self.tableWidget.verticalHeaderItem(2)
        item.setText(_translate("MainWindow", "New Row"))
        item = self.tableWidget.horizontalHeaderItem(0)
        item.setText(_translate("MainWindow", "Ocean Proximity Value"))
        item = self.tableWidget.horizontalHeaderItem(1)
        item.setText(_translate("MainWindow", "Age range"))
        item = self.tableWidget.horizontalHeaderItem(2)
        item.setText(_translate("MainWindow", "Count"))
        self.label.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\"><span style=\" font-size:36pt;\">Housing Median Age by Ocean Proximity</span></p></body></html>"))
        self.pushButton.setText(_translate("MainWindow", "Load"))
        self.pushButton_2.setText(_translate("MainWindow", "Back"))
        self.comboBox.setItemText(0, _translate("MainWindow", "<1H OCEAN"))
        self.comboBox.setItemText(1, _translate("MainWindow", "INLAND"))
        self.comboBox.setItemText(2, _translate("MainWindow", "ISLAND"))
        self.comboBox.setItemText(3, _translate("MainWindow", "NEAR BAY"))
        self.comboBox.setItemText(4, _translate("MainWindow", "NEAR OCEAN"))
        self.label_2.setText(_translate("MainWindow", "Ocean Proximity value"))
        
class Ui_Third(object):
    def OnClick5(self):
        self.window = QtWidgets.QMainWindow()
        self.ui= Ui_MainWindow()
        self.ui.setupUi(self.window)
        MainWindow.hide()
        self.window.show()
    def draw_line_graph(self):
        # Create a QGraphicsScene to hold the line graph
        scene = QtWidgets.QGraphicsScene()
        self.graphicsView.setScene(scene)
        selected_index = self.comboBox.currentIndex()
        selected_item = self.comboBox.itemText(selected_index)

        # Execute the SQL query to retrieve the data for the line graph
        query = """
            SELECT l.month, l.year, AVG(l.loan_amount) as avg_loan_amount
            FROM LoanSanction ls
            JOIN DimLoan l ON ls.loan_id = l.loan_id
            WHERE l.year = '{}'
            GROUP BY l.month, l.year
            ORDER BY l.year, FIELD(l.month, 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December')
        """.format(selected_item)

        # Replace the connection details with your own
        from DATA225utils import make_connection
        conn = make_connection('centaurus_wh.ini')
        cursor = conn.cursor()

        cursor.execute(query)
        data = cursor.fetchall()

        max_value = max(row[2] for row in data)

        # Create a QPen to draw the line
        pen = QtGui.QPen(QtGui.QColor(80, 150, 200))
        pen.setWidth(3)

        # Calculate the x-axis and y-axis scales
        x_scale = self.graphicsView.width() / (len(data) - 1)
        y_scale = (self.graphicsView.height() - 50) / max_value  # Subtract additional vertical space here

        # Create a QPointF for the first data point
        first_point = QtCore.QPointF(0, self.graphicsView.height() - data[0][2] * y_scale)

        # Loop through the data and add a line for each point
        for i, row in enumerate(data):
            # Calculate the position of the data point
            x = i * x_scale
            y = self.graphicsView.height() - row[2] * y_scale

            # Create a QGraphicsLineItem to represent the line segment
            line = QtWidgets.QGraphicsLineItem(first_point.x(), first_point.y(), x, y)
            line.setPen(pen)
            scene.addItem(line)

            # Update the first_point for the next line segment
            first_point = QtCore.QPointF(x, y)

            # Add a label for the data point
            label = QtWidgets.QGraphicsTextItem(f"{row[1]} {row[0]}")
            label.setPos(x - label.boundingRect().width() / 2, y - 20)
            scene.addItem(label)

        # Add x-axis label
        x_label = QtWidgets.QGraphicsTextItem("Month and Year")
        x_label.setPos(self.graphicsView.width() / 2 - x_label.boundingRect().width() / 2, self.graphicsView.height() + 10)
        scene.addItem(x_label)

        # Add y-axis label
        y_label = QtWidgets.QGraphicsTextItem("Average Loan Amount")
        y_label.setPos(-50, self.graphicsView.height() / 2 - y_label.boundingRect().height() / 2)
        y_label.setRotation(-90)
        scene.addItem(y_label)

        # Close the database connection
        conn.close()


        
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(1470, 836)
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
        self.label = QtWidgets.QLabel(self.centralwidget)
        self.label.setGeometry(QtCore.QRect(430, 60, 691, 51))
        self.label.setObjectName("label")
        self.graphicsView = QtWidgets.QGraphicsView(self.centralwidget)
        self.graphicsView.setGeometry(QtCore.QRect(100, 200, 711, 531))
        self.graphicsView.setObjectName("graphicsView")
        self.pushButton = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton.setGeometry(QtCore.QRect(990, 410, 113, 32))
        self.pushButton.setObjectName("pushButton")
        self.pushButton.clicked.connect(self.draw_line_graph)
        self.comboBox = QtWidgets.QComboBox(self.centralwidget)
        self.comboBox.setGeometry(QtCore.QRect(990, 350, 104, 26))
        self.comboBox.setObjectName("comboBox")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.comboBox.addItem("")
        self.label_2 = QtWidgets.QLabel(self.centralwidget)
        self.label_2.setGeometry(QtCore.QRect(900, 350, 60, 16))
        self.label_2.setObjectName("label_2")
        self.pushButton_2 = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton_2.setGeometry(QtCore.QRect(990, 480, 113, 32))
        self.pushButton_2.setObjectName("pushButton_2")
        self.pushButton_2.clicked.connect(self.OnClick5)
        MainWindow.setCentralWidget(self.centralwidget)
        self.menubar = QtWidgets.QMenuBar(MainWindow)
        self.menubar.setGeometry(QtCore.QRect(0, 0, 1470, 36))
        self.menubar.setObjectName("menubar")
        MainWindow.setMenuBar(self.menubar)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)

        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "MainWindow"))
        self.label.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\"><span style=\" font-size:36pt;\">Average Loan Amount by Month and Year</span></p></body></html>"))
        self.pushButton.setText(_translate("MainWindow", "Load"))
        self.comboBox.setItemText(0, _translate("MainWindow", "2000"))
        self.comboBox.setItemText(1, _translate("MainWindow", "2001"))
        self.comboBox.setItemText(2, _translate("MainWindow", "2002"))
        self.comboBox.setItemText(3, _translate("MainWindow", "2003"))
        self.comboBox.setItemText(4, _translate("MainWindow", "2004"))
        self.comboBox.setItemText(5, _translate("MainWindow", "2005"))
        self.comboBox.setItemText(6, _translate("MainWindow", "2006"))
        self.comboBox.setItemText(7, _translate("MainWindow", "2007"))
        self.comboBox.setItemText(8, _translate("MainWindow", "2008"))
        self.comboBox.setItemText(9, _translate("MainWindow", "2009"))
        self.comboBox.setItemText(10, _translate("MainWindow", "2010"))
        self.comboBox.setItemText(11, _translate("MainWindow", "2011"))
        self.comboBox.setItemText(12, _translate("MainWindow", "2012"))
        self.comboBox.setItemText(13, _translate("MainWindow", "2013"))
        self.comboBox.setItemText(14, _translate("MainWindow", "2014"))
        self.comboBox.setItemText(15, _translate("MainWindow", "2015"))
        self.label_2.setText(_translate("MainWindow", "year"))
        self.pushButton_2.setText(_translate("MainWindow", "Back"))
        
class Ui_Lastpage(object):
    def OnClick7(self):
        self.window = QtWidgets.QMainWindow()
        self.ui= Ui_MainWindow()
        self.ui.setupUi(self.window)
        MainWindow.hide()
        self.window.show()
        
    def draw_bar_graph(self):
        # Create a QGraphicsScene to hold the bar graph
        scene = QtWidgets.QGraphicsScene()
        self.graphicsView.setScene(scene)

        # Execute the SQL query to retrieve the data for the bar graph
            # Execute the SQL query to retrieve the data for the bar graph
        query = """
            SELECT dp.ocean_proximity_value, COUNT(*) AS num_houses
            FROM DimHousing AS dh
            JOIN HouseOwnership AS ho ON dh.housing_id = ho.housing_id
            JOIN DimOceanProximity AS dp ON ho.proximity_id = dp.proximity_id
            WHERE dh.median_house_value > 500000
            GROUP BY dp.ocean_proximity_value
            ORDER BY num_houses DESC
            LIMIT 10;
        """
        # Replace the connection details with your own
        from DATA225utils import make_connection
        conn = make_connection('centaurus_wh.ini')
        cursor = conn.cursor()

        cursor.execute(query)
        data = cursor.fetchall()

        max_value = max(row[1] for row in data)

        # Create a QBrush to fill the bars in the graph
        brush = QtGui.QBrush(QtGui.QColor(149, 175, 192))

        # Calculate the width of each bar
        bar_width = self.graphicsView.width() / len(data)

        # Loop through the data and add a bar for each value
        for i, row in enumerate(data):
            # Calculate the height of the bar
            bar_height = (row[1] / max_value) * self.graphicsView.height()

            # Create a QGraphicsRectItem to represent the bar
            rect = QtWidgets.QGraphicsRectItem(
                i * bar_width, self.graphicsView.height() - bar_height,
                bar_width, bar_height
            )

            # Set the brush for the bar
            rect.setBrush(brush)

            # Add the bar to the scene
            scene.addItem(rect)

            # Add the value label above the bar
            value_label = QtWidgets.QGraphicsTextItem(str(row[1]))
            value_label.setPos(i * bar_width + bar_width / 2 - value_label.boundingRect().width() / 2, self.graphicsView.height() - bar_height - 20)
            scene.addItem(value_label)

            # Add x-axis label
            x_label = QtWidgets.QGraphicsTextItem(row[0])
            x_label.setPos(i * bar_width + bar_width / 2 - x_label.boundingRect().width() / 2, self.graphicsView.height() + 10)
            scene.addItem(x_label)

        # Add y-axis label
        y_label = QtWidgets.QGraphicsTextItem("Number of Houses")
        y_label.setPos(-50, self.graphicsView.height() / 2 - y_label.boundingRect().height() / 2)
        y_label.setRotation(-90)
        scene.addItem(y_label)

        self.graphicsView.scale(0.8, 0.8)

        # Close the database connection
        conn.close()
        
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(1470, 834)
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
        self.label = QtWidgets.QLabel(self.centralwidget)
        self.label.setGeometry(QtCore.QRect(350, 50, 871, 61))
        self.label.setObjectName("label")
        self.graphicsView = QtWidgets.QGraphicsView(self.centralwidget)
        self.graphicsView.setGeometry(QtCore.QRect(100, 180, 721, 611))
        self.graphicsView.setObjectName("graphicsView")
        self.pushButton = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton.setGeometry(QtCore.QRect(990, 360, 113, 32))
        self.pushButton.setObjectName("pushButton")
        self.pushButton.clicked.connect(self.draw_bar_graph)
        self.pushButton_2 = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton_2.setGeometry(QtCore.QRect(990, 430, 113, 32))
        self.pushButton_2.setObjectName("pushButton_2")
        self.pushButton_2.clicked.connect(self.OnClick7)
        MainWindow.setCentralWidget(self.centralwidget)
        self.menubar = QtWidgets.QMenuBar(MainWindow)
        self.menubar.setGeometry(QtCore.QRect(0, 0, 1470, 36))
        self.menubar.setObjectName("menubar")
        MainWindow.setMenuBar(self.menubar)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)

        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "MainWindow"))
        self.label.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\"><span style=\" font-size:36pt;\">Ocean Proximity Values for High-Value Homes</span></p></body></html>"))
        self.pushButton.setText(_translate("MainWindow", "Load"))
        self.pushButton_2.setText(_translate("MainWindow", "Back"))


if __name__ == "__main__":
    import sys
    app = QtWidgets.QApplication(sys.argv)
    MainWindow = QtWidgets.QMainWindow()
    ui = Ui_MainWindow()
    ui.setupUi(MainWindow)
    MainWindow.show()
    sys.exit(app.exec_())
```
